```python
import pandas as pd
import numpy  as np
import seaborn as sns 
from sklearn.cluster import KMeans

pd.options.display.max_rows = 500000
```

```python
def Troca (file):
    
    S = []
    for i in range(len(file["Valor de Venda"])):
        if float==type(file["Valor de Venda"][i]):
            DT = file["Valor de Venda"][i]
        else:
            DT = file["Valor de Venda"][i].replace(',','.')
        S.append(float(DT))
    file["Valor de Venda(1)"] = S
```

```python

for d in range(2004,2023):
    
    name_=("{0}").format(d)
    file_name_1 = ("glp-{0}-01.csv").format(d)
    file_name_2 = ("glp-{0}-02.csv").format(d)
    file_1 = pd.read_csv(file_name_1,sep=';')
    file_2 = pd.read_csv(file_name_2,sep=';')
    Troca(file_1)
    Troca(file_2)
    file_3 = pd.merge(file_1.groupby("Estado - Sigla")["Valor de Venda(1)"].mean(),file_2.groupby("Estado - Sigla")["Valor de Venda(1)"].mean(), on="Estado - Sigla", suffixes=('', '1'))
    file_3 = pd.Series(file_3.mean(axis=1),name= name_)
    if(d == 2004):
        file_glp = file_3
    else:
        file_glp = pd.merge(file_glp , file_3, on="Estado - Sigla", suffixes=("",""))

file_glp
```
